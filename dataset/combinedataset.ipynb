```python
import os
import json
import pandas as pd

base = "APPS/APPS/test"

rows = []

problem_dirs = os.listdir(base)

for pid in problem_dirs:

    pdir = os.path.join(base, pid)

    try:
        question_path = os.path.join(pdir, "question.txt")
        solution_path = os.path.join(pdir, "solutions.json")
        io_path = os.path.join(pdir, "input_output.json")
        metadata_path = os.path.join(pdir, "metadata.json")

        with open(question_path, "r", encoding="utf-8") as f:
            question = f.read()

        with open(solution_path, "r", encoding="utf-8") as f:
            solutions = json.load(f)

        with open(io_path, "r", encoding="utf-8") as f:
            io_data = json.load(f)

        with open(metadata_path, "r", encoding="utf-8") as f:
            metadata = json.load(f)

        rows.append({
            "problem_id": pid,
            "question": question,
            "solutions": solutions,
            "input_output": io_data,
            "difficulty": metadata.get("difficulty", None),
            "url": metadata.get("url", None)
        })

    except Exception as e:
        print(f"Failed {pid}: {e}")

apps_df = pd.DataFrame(rows)

print(apps_df.head())

print(apps_df.columns)

print(len(apps_df))
```

```python
import os
import json
import pandas as pd

# =========================================================
# GLOBAL CONFIG
# =========================================================

VALID_EXTENSIONS = [
    ".py",
    ".cpp",
    ".c",
    ".js",
    ".java",
    ".ts"
]

IGNORE_DIRS = [
    ".git",
    "__pycache__",
    "node_modules",
    "build",
    "dist",
    ".pytest_cache",
    "venv",
    ".idea",
    ".vscode"
]

MAX_FILE_SIZE = 200_000

IMPORTANT_FILES = [
    "PRD.md",
    "architecture_design.md",
    "repo_config.json"
]

# =========================================================
# SAFE FILE READER
# =========================================================

def safe_read(path):

    try:

        if os.path.getsize(path) > MAX_FILE_SIZE:
            return None

        with open(path, "r", encoding="utf-8") as f:
            return f.read()

    except:
        return None


# =========================================================
# SWE-BENCH CONVERTER
# =========================================================

def convert_swebench(df):

    rows = []

    for _, row in df.iterrows():

        instruction = (
            "Fix the bug described in the repository issue."
        )

        inp = f"""
Repository:
{row['repo']}

Problem Statement:
{row['problem_statement']}

Hints:
{row['hints_text']}
"""

        output = row["patch"]

        rows.append({
            "source": "swebench",
            "task_type": "bug_fixing",
            "instruction": instruction,
            "input": inp,
            "output": output,
            "metadata": {
                "instance_id": row["instance_id"]
            }
        })

    return rows


# =========================================================
# CLASSEVAL CONVERTER
# =========================================================

def convert_classeval(df):

    rows = []

    for _, row in df.iterrows():

        instruction = (
            "Complete the Python class implementation."
        )

        inp = f"""
Class Description:
{row['class_description']}

Skeleton:
{row['skeleton']}

Tests:
{row['test']}
"""

        output = row["solution_code"]

        rows.append({
            "source": "classeval",
            "task_type": "class_implementation",
            "instruction": instruction,
            "input": inp,
            "output": output,
            "metadata": {
                "class_name": row["class_name"]
            }
        })

    return rows


# =========================================================
# APPS CONVERTER
# =========================================================

def convert_apps(df):

    rows = []

    for _, row in df.iterrows():

        if row["difficulty"] == "introductory":
            continue

        solutions = row["solutions"]

        if not solutions:
            continue

        output = solutions[0]

        rows.append({
            "source": "apps",
            "task_type": "algorithmic_reasoning",
            "instruction": "Solve the programming problem.",
            "input": row["question"],
            "output": output,
            "metadata": {
                "difficulty": row["difficulty"]
            }
        })

    return rows


# =========================================================
# DEVBENCH PARSER
# =========================================================

def parse_devbench_project(project_path):

    data = {
        "project": os.path.basename(project_path),
        "prd": "",
        "architecture": "",
        "tests": [],
        "source": []
    }

    for root, dirs, files in os.walk(project_path):

        dirs[:] = [d for d in dirs if d not in IGNORE_DIRS]

        for file in files:

            full = os.path.join(root, file)

            # -----------------------------------
            # IMPORTANT DOC FILES
            # -----------------------------------

            if file in IMPORTANT_FILES:

                content = safe_read(full)

                if content:

                    if file == "PRD.md":
                        data["prd"] = content

                    elif file == "architecture_design.md":
                        data["architecture"] = content

            # -----------------------------------
            # TEST FILES
            # -----------------------------------

            elif "test" in file.lower():

                content = safe_read(full)

                if content:
                    data["tests"].append(content)

            # -----------------------------------
            # SOURCE FILES
            # -----------------------------------

            else:

                ext = os.path.splitext(file)[1]

                if ext in VALID_EXTENSIONS:

                    content = safe_read(full)

                    if content:
                        data["source"].append(content)

    return data


# =========================================================
# DEVBENCH CONVERTER
# =========================================================

def convert_devbench(devbench_base):

    rows = []

    for language in os.listdir(devbench_base):

        lang_path = os.path.join(devbench_base, language)

        if not os.path.isdir(lang_path):
            continue

        for project in os.listdir(lang_path):

            project_path = os.path.join(lang_path, project)

            if not os.path.isdir(project_path):
                continue

            print(f"Parsing: {project}")

            parsed = parse_devbench_project(project_path)

            # -----------------------------------
            # ARCHITECTURE TASK
            # -----------------------------------

            if parsed["prd"] and parsed["architecture"]:

                rows.append({

                    "source": "devbench",

                    "task_type": "architecture_reasoning",

                    "instruction":
                        "Design a production-ready architecture for the project.",

                    "input":
                        parsed["prd"],

                    "output":
                        parsed["architecture"],

                    "metadata": {
                        "project": parsed["project"],
                        "language": language
                    }
                })

            # -----------------------------------
            # IMPLEMENTATION TASK
            # -----------------------------------

            if parsed["prd"] and parsed["source"]:

                source_code = "\n\n".join(parsed["source"][:5])

                tests = "\n\n".join(parsed["tests"][:3])

                rows.append({

                    "source": "devbench",

                    "task_type": "implementation",

                    "instruction":
                        "Implement the project from the product requirements.",

                    "input":
                        parsed["prd"] + "\n\n" + tests,

                    "output":
                        source_code,

                    "metadata": {
                        "project": parsed["project"],
                        "language": language
                    }
                })

    return rows


# =========================================================
# BUILD ALL DATA
# =========================================================

all_rows = []

# SWE-BENCH
all_rows.extend(convert_swebench(SWEdf))

# CLASSEVAL
all_rows.extend(convert_classeval(FSEtrain_df))

# APPS
all_rows.extend(convert_apps(apps_df))

# DEVBENCH
devbench_rows = convert_devbench("../DevBench/benchmark_data")
all_rows.extend(devbench_rows)

# =========================================================
# FINAL DATAFRAME
# =========================================================

combined_df = pd.DataFrame(all_rows)

# REMOVE DUPLICATES
combined_df = combined_df.drop_duplicates(
    subset=["instruction", "input", "output"]
)

# REMOVE EMPTY OUTPUTS
combined_df = combined_df[
    combined_df["output"].str.len() > 20
]

print("=" * 60)
print("FINAL DATASET SIZE")
print(len(combined_df))
print("=" * 60)

print(combined_df.head())

# =========================================================
# SAVE JSONL
# =========================================================

combined_df.to_json(
    "qforge_dataset.jsonl",
    orient="records",
    lines=True
)

print("Saved -> qforge_dataset.jsonl")


```

```python
import json
import hashlib
import random
from collections import Counter

# ============================================================
# CONFIG
# ============================================================

INPUT_FILE = "qforge_dataset.jsonl"

OUTPUT_CLEAN = "qforge_cleaned.jsonl"

MAX_INPUT_CHARS = 12000
MAX_OUTPUT_CHARS = 12000

MIN_OUTPUT_CHARS = 20

# ============================================================
# LOAD DATASET
# ============================================================

rows = []

with open(INPUT_FILE, "r", encoding="utf-8") as f:

    for line in f:

        try:
            rows.append(json.loads(line))

        except Exception as e:
            print("BROKEN JSON:", e)

print("=" * 80)
print("RAW DATASET SIZE:", len(rows))

# ============================================================
# SHOW RAW SAMPLE
# ============================================================

print("\nRAW SAMPLE:\n")

for x in random.sample(rows, 2):

    print("=" * 80)

    print(json.dumps(x, indent=2)[:3000])

# ============================================================
# CLEANING HELPERS
# ============================================================

def normalize_text(x):

    if x is None:
        return ""

    x = str(x)

    x = x.replace("\x00", "")
    x = x.replace("\r\n", "\n")
    x = x.strip()

    return x


def sample_hash(sample):

    text = (
        sample["instruction"]
        + sample["input"]
        + sample["output"]
    )

    return hashlib.md5(text.encode()).hexdigest()


def looks_bad_output(text):

    text_lower = text.lower()

    BAD_PATTERNS = [
        "as an ai language model",
        "i cannot assist",
        "i'm sorry but",
        "openai policy",
        "ethical guidelines"
    ]

    for p in BAD_PATTERNS:
        if p in text_lower:
            return True

    return False


# ============================================================
# CLEAN DATASET
# ============================================================

cleaned = []

seen = set()

removed_stats = Counter()

for row in rows:

    try:

        instruction = normalize_text(
            row.get("instruction", "")
        )

        inp = normalize_text(
            row.get("input", "")
        )

        output = normalize_text(
            row.get("output", "")
        )

        source = normalize_text(
            row.get("source", "unknown")
        )

        task_type = normalize_text(
            row.get("task_type", "unknown")
        )

        # ----------------------------------------------------
        # EMPTY FILTER
        # ----------------------------------------------------

        if not instruction:
            removed_stats["empty_instruction"] += 1
            continue

        if not output:
            removed_stats["empty_output"] += 1
            continue

        # ----------------------------------------------------
        # TOO SHORT
        # ----------------------------------------------------

        if len(output) < MIN_OUTPUT_CHARS:
            removed_stats["short_output"] += 1
            continue

        # ----------------------------------------------------
        # TOO LARGE
        # ----------------------------------------------------

        if len(inp) > MAX_INPUT_CHARS:
            removed_stats["large_input"] += 1
            continue

        if len(output) > MAX_OUTPUT_CHARS:
            removed_stats["large_output"] += 1
            continue

        # ----------------------------------------------------
        # BAD REFUSAL FILTER
        # ----------------------------------------------------

        if looks_bad_output(output):
            removed_stats["refusal"] += 1
            continue

        # ----------------------------------------------------
        # FINAL SAMPLE
        # ----------------------------------------------------

        sample = {
            "source": source,
            "task_type": task_type,
            "instruction": instruction,
            "input": inp,
            "output": output
        }

        # ----------------------------------------------------
        # DEDUPLICATION
        # ----------------------------------------------------

        h = sample_hash(sample)

        if h in seen:
            removed_stats["duplicate"] += 1
            continue

        seen.add(h)

        cleaned.append(sample)

    except Exception as e:

        removed_stats["failed"] += 1

        print("FAILED ROW:", e)

# ============================================================
# RESULTS
# ============================================================

print("\n" + "=" * 80)

print("CLEANED DATASET SIZE:", len(cleaned))

print("\nREMOVAL STATS:\n")

for k, v in removed_stats.items():
    print(f"{k}: {v}")

# ============================================================
# SHOW CLEANED SAMPLES
# ============================================================

print("\nCLEANED SAMPLES:\n")

for x in random.sample(cleaned, 3):

    print("=" * 80)

    print(json.dumps(x, indent=2)[:4000])

# ============================================================
# TASK DISTRIBUTION
# ============================================================

task_counts = Counter()

for row in cleaned:
    task_counts[row["task_type"]] += 1

print("\nTASK DISTRIBUTION:\n")

for k, v in task_counts.items():
    print(k, ":", v)

# ============================================================
# SAVE CLEANED DATASET
# ============================================================

with open(OUTPUT_CLEAN, "w", encoding="utf-8") as f:

    for row in cleaned:

        f.write(
            json.dumps(row, ensure_ascii=False) + "\n"
        )

print("\nSAVED:", OUTPUT_CLEAN)
```
